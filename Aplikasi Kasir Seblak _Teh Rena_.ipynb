```python
menus = [
    {
        "name" : "kerupuk merah",
        "price" : 2000
    },
    {
        "name" : "kerupuk bunga",
        "price" : 2000
    },
    {
        "name" : "kwetiauw",
        "price" : 3000
    },
    {
        "name" : "makaroni",
        "price" : 2500
    },
    {
        "name" : "mie kuning",
        "price" : 1500
    },
    {
        "name" : "telor",
        "price" : 2000
    },
    {
        "name" : "sosis",
        "price" : 2500
    },
    {
        "name" : "jamur enoki",
        "price" : 4000
    }
]
Varian_kuah = [
    {
        "kuah" : "original",
        "price" : 1500
    },
    {
        "kuah" : "tom yam",
        "price" : 1500
    }
]
nama = input("input nomer antrian : ")
topping_seblak_dipesan = []
topping_pesanan_all = []
total_harga = 0

for item in menus:
  topping_pesanan_all.append(item["name"])

minimal_topping = 0
while True:
  pesanan = input("input topping seblak di sini:")
  if pesanan in topping_pesanan_all:
    topping_seblak_dipesan.append(pesanan)
    for item in menus:
      if pesanan == item["name"]:
        total_harga += item["price"]
        minimal_topping += 1
  else:
    print("pesanan tidak tersedia")
  if minimal_topping >= 3:
    break
  else:
    print("minimal harus 3 topping")
    print("masukan lagi")
    continue
  print("Apakah sudah selesai?")
  done = input()
  if done == "ya":
      break

varian_kuah_dipesan =[]
varian_kuah_all=[]
for variant_kuah in Varian_kuah:
  varian_kuah_all.append(variant_kuah["kuah"])
while True:
  pesanan = input("input varian kuah di sini:")
  if pesanan in varian_kuah_all:
    varian_kuah_dipesan.append(pesanan)
    for item in Varian_kuah:
      if pesanan == item["kuah"]:
        total_harga += item["price"]
  else:
    print("pesanan tidak tersedia")
  print("Apakah sudah selesai?")
  done = input()

  if done == "ya":
      break
print(topping_seblak_dipesan)
print(varian_kuah_dipesan)
print("total harga:",total_harga)
```
